```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df=pd.read_csv('train.csv')
```

```python
train_df.head()
```

```python
train_df=train_df.drop(['PassengerId'],axis=1)
```

# Exploratory Data Analysis

## Gender

```python
sns.countplot(x='Sex',hue='Survived',data=train_df)
```

## Age

```python
sns.displot(train_df, x="Age", hue="Survived",kind='kde')
```

## Ticket Class

```python
sns.countplot(x='Pclass',hue='Survived',data=train_df)
```

## Fare

```python
sns.displot(train_df, x="Fare", hue="Survived",kind='kde')
```

## Port of Embarkation

```python
sns.countplot(x='Embarked',hue='Survived',data=train_df)
```

## Number of sibilings and spouses onboard

```python
sns.countplot(x='SibSp',hue='Survived',data=train_df)
```

## Number of parents/children onboard

```python
sns.countplot(x='Parch',hue='Survived',data=train_df)
```

## Correlation Matrix

```python
corr_mat = train_df.corr(numeric_only=True)
sns.heatmap(corr_mat, cmap="YlGnBu", annot=True)
```

From this we can observe that the largest correlations with whether a passenger surived is class of ticket and fare price.
